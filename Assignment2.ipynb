```python
#importing necessary libraries
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

```python
#import dataset and split into train and test data
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
plt.matshow(x_train[1])
```

```python
# This line displays the inverted grayscale image of the first sample in x_train using Matplotlib
plt.imshow(-x_train[0], cmap="gray")
```

```python
x_train = x_train / 255
x_test = x_test / 255
```

```python
# Creates a sequential model to stack layers linearly.
# This means each layer’s output becomes the input to the next layer in the sequence
model = keras.Sequential([
# Flattens the 28x28 input image into a 784-dimensional vector.
# flattening means converting a multi-dimensional array (like a 2D image) into a 1D array.
keras.layers.Flatten(input_shape=(28, 28)),
# Dense layer (also called a fully connected layer) means that each neuron (node) in this layer is connected to every neuron in the previous layer.
keras.layers.Dense(128, activation="relu"),
keras.layers.Dense(10, activation="softmax")
])

model.summary()
```

```python
model.compile(optimizer="sgd", #sdg used to minimize the loss function(A loss function is a measure of how well a machine learning model's predictions match the actual data.)
loss="sparse_categorical_crossentropy",
metrics=['accuracy'])
```

```python
# By saving history, you can later analyze the learning trend
history=model.fit(x_train,
y_train,validation_data=(x_test,y_test),epochs=10)
```

```python
test_loss,test_acc=model.evaluate(x_test,y_test)
print("Loss=%.3f" %test_loss)
print("Accuracy=%.3f" %test_acc)
```
